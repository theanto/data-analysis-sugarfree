```python
import pandas as pd 
import numpy as np

data = pd.read_csv("food.json") 
```

```python
data
```

```python


data.loc[data['idMeal'] == 1, 'time'] = '08:00:00'
data.loc[data['idMeal'] == 2, 'time'] = '10:00:00'
data.loc[data['idMeal'] == 3, 'time'] = '13:00:00'
data.loc[data['idMeal'] == 4, 'time'] = '16:00:00'
data.loc[data['idMeal'] == 5, 'time'] = '20:00:00'
data.loc[data['idMeal'] == 6, 'time'] = '22:00:00'


```

```python
data
```

```python
data["time"] = data["data"] +" "+ data["time"]
```

```python
data
```

```python
del data['data']
del data['logId']
del data['idMeal']
```

```python
data
```

```python
data = data.groupby(['uid','time'],as_index=False).agg({  'cal': 'sum', 'carbs': 'sum', 'fat': 'sum', 'fiber': 'sum', 'protein': 'sum', 'sodium': 'sum'})
```

```python
data
```

```python
export_csv = data.to_csv (r'food.csv', index = None, header=True)
```
