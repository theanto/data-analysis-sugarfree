```python
import pandas as pd 

data = pd.read_csv("workout.json", parse_dates=['time'] ) 
```

```python
data
```

```python
#data = data.set_index(['time'])
```

```python
dataM =data.reset_index().set_index('time')
```

```python
dataM
```

```python
dataM= dataM.fillna(1)
```

```python
dataM
```

```python
data =dataM.groupby('uid').resample('15min').mean().dropna()
```
