## Rate of change on SugarFree dataset

```python
%run -i "scriptROC.py"
```

```python
data= d[1]
```

```python
data
```

```python
data= data.set_index(['time'], inplace=False)
```

```python
l= [([0])]


for y in range(0, len(data)-1):



    val1 = data['sugarValue'][y+1:y+2].values
    val2 = data['sugarValue'][y:y+1].values


    l.append( (((val1 / val2 )-1)*100) )


data['dataROC'] = np.array(l)
```

```python

plotROC(data, "user 1")
```

```python
dataagg
```

```python
aggr = pd.DataFrame(columns=['time',  'dataROC'])

dataagg =data.reset_index()

for x in range(0, len(data)-3, 3):
   
    val = dataagg.iloc[x:x+3]


    #print(val["dataROC"])
    #print(val["dataROC"].mean())
    
    #print(val["timestamp"].iloc[2:3].values[0])
    time= pd.to_datetime(val["time"].iloc[2:3].values[0], format='%Y-%m-%d %H:%M:%S') 
    
    
    
    aggr = aggr.append({'time': time , 
                            'dataROC': val["dataROC"].mean()},ignore_index=True)
```

```python

aggr= aggr.set_index(['time'], inplace=False)
```

```python


plotROC(aggr, "user 1 average 15min")
```
